1. Certains des tweets ont été écrits en anglais, cependant, le groupe de tweets écrits en anglais par rapport aux tweets en arabe et en français était petit, à tel point que le temps et les ressources investis dans le traitement des données écrites en anglais n'étaient pas justifiables. Par conséquent, ces tweets ont été rejetés.
2. Bien que les tweets rédigés en langue arabe aient constitué l'essentiel des données disponibles, une partie considérable d'entre eux ont été rédigés en dialogique algérien par opposition à la langue arabe formelle. Cette découverte pose un défi, car les boîtes à outils en langage naturel (NLTK) disponibles ne peuvent pas traiter le contenu écrit en dialectes. Pour ce faire, un dictionnaire contenant tous les mots utilisés, leur racine, leur origine et un large corpus de textes écrits en dialecte devrait être fourni pour rendre cette entreprise possible. Cela nécessite des années de collecte et de traitement de données, ce qui est évidemment impossible dans notre cas.
Ces défis ne nous laissaient qu'une seule alternative, soit le traitement de données écrites en français. Suite à cette décision, 2475 tweets écrits en français contenant le hashtag `#oran2022`, ont été collectés <sup>[2](#fn2)</sup>.

La fonction `batch_collector()` prend une requête et renvoie un dictionnaire de dictionnaires. Un dictionnaire est une structure de données une classe de données comme une chaîne, un entier, une liste, etc. cela dépendrait de l'objet dans lequel nous voulons le mettre:

- Le type le plus basique est `string` : `"X a 40 ans et pèse 70kg"` ;
- Un objet `list` prendrait la forme suivante : `['X', '40', '70']` ;
- Un objet `dict` prendrait la forme suivante : `{'name' : 'X', 'age' : 40, 'weight' : 70}`.

Comme nous pouvons le voir, un objet `dict` est le plus descriptif des objets de base, ce qui le rend facile à lire et à comprendre pour quelqu'un qui n'a pas participé à la création des données. Bien que les développeurs soient familiers avec les structures de données de base, y compris l'objet `dict` ou son équivalent dans d'autres langages de programmation comme `map()` dans les langages `dart` et `javascript`. Notre fonction `batch_collecter()` renvoie un dictionnaire de dictionnaires, c'est-à-dire un dictionnaire contenant plusieurs dictionnaires et chacun de ces dictionnaires internes contient les données d'un seul tweet (par exemple, `nom d'utilisateur` de l'affiche, `contenu` du tweet, `date ` il a été tweeté, `liens` ou URL s'il y en a, nombre de likes et de citations `n_likes` & `n_quotes`, etc. ; voir les lignes 30 à 43 du bloc de code ci-dessous.

````python

time_start = datetime.now()
# a function to extract urls shared in a tweet
def get_listed_url(tweet_object):
    if tweet_object.links == None:
        return None
    elif len(tweet_object.links) == 1:
        return tweet_object.links[0].url
    elif len(tweet_object.links) > 1:
        return [u.url for u in tweet_object.links]
    else:
        return None
# the main function that collects the whole database
# receives a query, formulated using Twitter's advanced search form
# link: https://twitter.com/search-advanced
def batch_collector(query):
    # initiating an empty dictionary to be populated by data collected
    # raw data is cleaned and formatted
    # only certain entries are of interest, i.e.: 
    # username, content, date, tweet_id, outlinks in case a user shares a link in their tweet, \
    # number of links shared, number of times tweet has been quoted, \
    # number of times tweet has been liked, the language (though in this example only French was selected) \
    # number of replies to the tweet, list of users quoting tweet, list of users liking tweet.
    tweets_dict = dict()
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(query).get_items()):
        # Sets a limit for the maximum tweets to be collected
        if i >= 5000:
            break
        
        tweets_dict.update({i:{
            'username': tweet.user.username,
            'content':tweet.rawContent, 
            'date':tweet.date, 
            'tweet_id':tweet.id, 
            'outlinks': get_listed_url(tweet),
            'n_links': len(tweet.links) if tweet.links != None else 0,
            'url':tweet.url,
            'n_likes':tweet.likeCount,
            'n_quotes': tweet.quoteCount,
            'language': tweet.lang,
            'replies': tweet.replyCount,
            'tweet_quotes': np.nan,
            'tweet_likers': np.nan,
         }})
    return tweets_dict

twittools.save_to_csv(df, 'oran_22.csv')
twittools.save_to_excel(df, 'oran_22.xlsx')
````

Le bloc de code ci-dessous se compose de deux fonctions qui font partie du module `twittools`. Ces deux fonctions suppriment les liens, les hashtags et les mentions.

<a name="fn11">11</a> Lectures complémentaires sur la tokenisation, la lemmatisation : https://medium.com/swlh/tweets-classification-and-clustering-in-python-b107be1ba7c7

````python
para = "loram ipsum @dolor #adisipin asit amet"
para_list = para.split(' ')
raw_tweet = str()


# this is a function to extract: hashtags, @mentions, and links
def tweet_parser(tweet):
    # URL pattern for url detection in tweet
    re_expression = '[(http(s)?):\/\/(www\.)?a-zA-Z0-9@:%._\+~#=]{2,256}\.[a-zأ-ي]{2,6}\b([-a-zA-Zأ-ي0-9@:%_\+.~#?&//=]*)'
    # replacing any url with ''
    tweet = re.sub(re_expression, '', tweet)
    # initiating empty list for hashtags
    hashtags = list()
    # initiating empty list for mentions
    mentions = list()
    tweet_to_list = tweet.split(' ')
    for i, word in enumerate(tweet_to_list):
        if (word.startswith('#') or word.endswith('#')):
            hashtags.append(word)
        elif (word.startswith('@') or word.endswith('@')):
            mentions.append(word)
            
    return hashtags if hashtags else None
    
def remove_hashtags(tweet):
    tweet = tweet.replace('\n', ' ')
    tweet_to_list = tweet.split(' ')
    for i, word in enumerate(tweet_to_list):
        if (word.startswith('#') or word.endswith('#')):
            tweet_to_list.pop(i)
        elif (word.startswith('@') or word.endswith('@')):
            tweet_to_list.pop(i)
    return ' '.join(tweet_to_list) if tweet_to_list else None
````

Ci-dessous se trouve une fonction `emoji_parser()` qui fait partie du module `twittools` qui prend un texte brut et renvoie le texte sans emojis ainsi qu'une liste d'emojis (voir sortie ci-dessous).

````python

# Function to parse tweets, returns raw text (emojiless), and a list of emojis
def emoji_parser(text):
    emoji_list = []
    # \X matches any unicode grapheme, e.g., ©
    # Note: here "regex" library is used which is external; and not "re" library
    # data captures all graphemes, but not all graphemes are emojis
    data = regex.findall(r'\X', text)
    # data needs to be cleaned using all recognized emojis filter list (.EMOJI_DATA)
    for word in data:
        # if there is a match, the grapheme is appended to a list of emojis
        if any(char in emoji.unicode_codes.EMOJI_DATA for char in word):
            emoji_list.append(word)
    if emoji_list:
        for e in emoji_list:
            text = text.replace(e, '')
    flags = [flag for flag in twittools.flags if flag in text]
    emoji_list = flags + emoji_list
    # this is a parser that removes any unicode characters pointing to flags e.g., '🇩🇿'.
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('ascii')
    # returns emojiless string and a list of emojis, e.g., ('lorem ipsum dolor!', ['👍', '❤️'])
    return text, emoji_list


text = "👍Lorem ipsum dolor!🇩🇿👍❤️"
emoji_parser(text)
````

Après avoir transformé tous nos documents texte en minuscules, nous effectuons les étapes de tokenisation `.tokenize` et de lemmatisation `lemmatize` en même temps avec la fonction `furnished()` qui fait partie du module `twittools`, nous avons ensuite créé un nouveau colonne `bag_of_words` qui est juste un sac de mots où l'ordre n'a plus d'importance, bien que dans certains cas l'ordre des mots puisse offrir plus d'informations, c'est dans ce but que nous avons décidé de conserver l'ordre des mots.

````python
stop = list(stopwords.words('french'))
stop = stop + ['http', 'https', 'co', 'avoir', 'tout', 'cad']
stop = set(stop)
tokenizer = RegexpTokenizer(r'\w+')
nlp = spacy.load('fr_core_news_md')
punctuation = list(string.punctuation)
stop.update(punctuation)
w_tokenizer = WhitespaceTokenizer()
def furnished(tweet):
    final_text = []
    for i_token in w_tokenizer.tokenize(tweet):
       if i_token.lower() not in stop:
           word = nlp(u"%s" %i_token)
           final_text.append(word[0].lemma_.lower())
    return " ".join(final_text)
````

Nous appliquons la fonction `furnished()` à chaque ligne de la colonne `content` et plaçons le sac de mots renvoyé dans la cellule correspondante sous la colonne `bag_of_words`.

````python
df2['c_bag_of_words'] = df2['content'].apply(furnished)
df2
````

Nous allons travailler avec du clustering 2D, c'est-à-dire un clustering entre 2 variables. Il existe différentes méthodes pour déterminer le nombre optimal de clusters et l'une d'entre elles est la méthode du *coude* de l'anglais *elbow method*. L'approche consiste à rechercher un coude dans le graphe WCSS, le coude dans le graph ci-dessous se trouve dans le cluster `x=3`. Habituellement, la partie du graphique avant le coude serait en forte baisse, tandis que la partie après serait plus lisse.

La méthode du coude est similaire à celle de l'analyse en composantes principales en statistique qui vise à identifier des groupes homogènes de variables. L'analyse de cluster vise à minimiser les distances entre les points/variables d'un cluster donné et à maximiser les distances entre les clusters. En règle générale, il peut y avoir 1 à n clusters possibles, parmi les moyens d'identifier le nombre optimal de clusters, il y a les méthodes du coude.

![alt text](elbow_method.png "Exemple de méthode du coude")

Typiquement, pour déterminer le nombre optimal de clusters, nous devons sélectionner la valeur de k au "coude", c'est-à-dire le point après lequel la distorsion ou l'inertie (c'est-à-dire la propagation ou la variance autour de la moyenne) commence à diminuer de manière linéaire .

En réalité cependant, ce n'est jamais aussi simple. Comme nous pouvons le voir sur notre graphique k-means, le coude n'est pas facile à discerner, il n'y a pas de *bonne réponse*, nous avons choisi arbitrairement 4 clusters, bien que choisir 6 ou 8 clusters aurait été tout aussi courant.

````python
wcss = []
corpus = df2['c_bag_of_words'].values.astype('U')
vectorizer = TfidfVectorizer(analyzer='word', stop_words=stop)
features = vectorizer.fit_transform(corpus)
for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0)
    kmeans.fit(features)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,15), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
````

Nous commençons maintenant l'analyse de clustering de texte avec le paramètre `n_clusters` égal à 4. Les autres paramètres n'ont aucune influence sur les résultats et sortent du cadre de ce document, nous n'en parlerons donc pas ici, nous fournissons cependant la documentation officielle de la classe `KMeans()` si le chercheur/praticien souhaite en savoir plus<sup>[13](#fn13)</sup>.

Suite à cette étape, nous créons une variable (colonne) nommée `cluster` où nous ajoutons à chaque tweet le cluster auquel ils appartiennent, cette colonne prend en charge les valeurs numériques, c'est-à-dire que `0` fait référence à `cluster_0`, `1` à `cluster_1` , etc.

<a name="fn13">13</a> Tous les paramètres pris en charge par la classe `KMeans()` expliqués (documentation) : https://scikit-learn.org/stable/modules/generated/sklearn.cluster. KMeans.html

````python
kmeans = KMeans(n_clusters=4, init='k-means++', n_init=10, max_iter=300, random_state=0)
Y_kmeans = kmeans.fit(features)
Y_kmeans.labels_
df2['cluster'] = Y_kmeans.labels_
df2['cluster'].value_counts()
save_to_csv(df2, 'oran_22_populated.csv')
save_to_excel(df2, 'oran_22_human_format_populated.xlsx')
````

Normaliser les données consiste à transformer nos données en pourcentages de la valeur moyenne pour réduire la distorsion et améliorer la lisibilité du graphique.

````python
cluster_labels = {
    0: 'individual sports / scores',
    1: 'opening ceremony / Moroccan delegation arrival',             
    2: 'Algerian-Moroccan politics',
    3: 'collective sports / group draw',
}
period_group = df.groupby(['period'])

x = period_group['cluster'].value_counts()
print(x)
pre = [0, 427, 194, 0]
game = [68, 1352, 206, 76]
post = [0, 120, 28, 4]
matrix = [pre, game, post]
matrix
matrix_transpose = list(zip(*matrix))
def normalize_list_of_lists(matrix):
    norm_matrix = []
    for i, l in enumerate(matrix):
        inner_row = [round((e/sum(l))*100, 1) for e in l]
        norm_matrix.append(inner_row)

    return norm_matrix
normalized_transpose_matrix = normalize_list_of_lists(matrix_transpose)
# print(l_2d_t_tuple)
# print(type(l_2d_t_tuple))
normalized_transpose_matrix
````
